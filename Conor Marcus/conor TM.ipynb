```python
#  instantiating BERTopic. We set language to multilingual




from bertopic import BERTopic

topic_model = BERTopic(language="multilingual", calculate_probabilities=False, verbose=True)
topics = topic_model.fit_transform(text_list)
```

To visually represent the decline in c-TF-IDF scores as words are added to a topic representation, consider the following approach:

Topics are characterized by a set of words, with each word assigned a c-TF-IDF score that reflects its importance to the topic. These scores gradually decrease as words are added to the topic representation. Eventually, adding more words to the topic provides only marginal improvements in the total c-TF-IDF score, making it less beneficial for topic representation.

To depict this phenomenon visually, we can create a plot that illustrates c-TF-IDF scores for each topic, organized by the rank of each word. Specifically, we will place the word's position (term rank) on the x-axis, where the words with the highest c-TF-IDF score receive a rank of 1. The y-axis will display the corresponding c-TF-IDF scores. The resulting visualization will demonstrate the decline in c-TF-IDF scores as words are incorporated into the topic representation. Utilizing the elbow method, you can then determine the optimal number of words to include in a topic, as it signifies the point at which further additions yield diminishing returns.

```python
topic_model.visualize_term_rank()
```
